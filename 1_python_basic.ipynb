- 멤버연산자 : 특정 데이터가 있는지(멤버인지) 확인 => not in / in

## Class

- 변수와 함수를 묶어 놓은 개념
- 사용 방법
    - 변수와 함수가 들어있는 클래스를 선언
    - 클래스를 객체로 만들어서 클래스 안에 선언된 변수와 함수를 사용
- 클래스 선언 시, 식별자는 camel case (변수, 함수는 snake case)
- 클래스 내 메소드 생성 시, 항상 self로 시작

````python
# 클래스의 선언
class Calculator:

    num1 = 1
    num2 = 2

    def __init__(self, name):
      self.method = name
      # self.num1 = 1
      # self.num2 = 2

    def plus(self):
        return self.num1+self.num2

    def minus(self):
        return self.num1-self.num2

#object 선언
calc = Calculator(name = '계산기')

print(calc.num1)
print(calc.num2)
print(calc.plus())
print(calc.minus())
print(calc.method)
````

### self의 의미 : 객체 자신

- '실제 세계' : 설계도(class)를 작성 -> 실제 물건(object)을 만들어냄.
